```python
class Chocolate:
  def __init__(self, price: float, adding: str = None): #Батончик (Цена, Добавка)
    if price > 0:
      self.price = price
      self.adding = adding
    else:
      raise ValueError('Цена не может быть отрицательной')
  def show_my_chocolate(self):
    if self.adding is not None:
      print(f'Шоколадный батончик, добавка: {self.adding}')
    else:
      print(f'Обычный шоколадный батончик')
  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'

choco = Chocolate(15.0, 'орехи')
choco.show_my_chocolate()
print(choco)
choco1 = Chocolate(5, 'карамель')
choco1.show_my_chocolate()
print(choco1)
choco2 = Chocolate(3.3)
choco2.show_my_chocolate()
print(choco2)
choco3 = Chocolate(-1.5)
```
