# Работа с API

```python
import requests
import pandas as pd
```

Согласно документации API root-ссылка:

```python
base_url = "https://www.anapioficeandfire.com/api"
```

Получаем информацию обо всех книгах и сохраняем в датафрейм PDbooks

```python
allbooks = requests.get(f'{base_url}/books').json()
```

```python
PDbooks = pd.DataFrame(allbooks)
```

```python
pd.set_option('display.max_colwidth', 400)
```

```python
PDbooks.head()
```

Получаем информацию о домах Вестероса и сохраняем в датафрейм PDhouses

```python
PDhouses = pd.DataFrame(requests.get(f'{base_url}/houses').json())
```

```python
PDhouses.head()
```

Получаем информацию о домах Вестероса с девизом и сохраняем в датафрейм PDhousesWithTitles

```python
PDhousesWithTitles = pd.DataFrame(requests.get(f'{base_url}/houses', params = {'hasTitles' : True}).json())
```

```python
PDhousesWithTitles.head()
```
